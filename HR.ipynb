Importing Dependencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
```

Data Collection and Processing

```python
# Never print the matching warning
warnings.filterwarnings('ignore')

# load the data from csv file to Pandas DataFrame
hr_data = pd.read_csv('/content/HR - Dataset.csv')

# printing the first 5 rows of the dataframe
hr_data.head()
```

```python
# Getting some informantion about data
hr_data.info()
```

```python
# Check the number of missing values in each column
hr_data.isnull().sum()
```

Data Analysis

Data Visualization

```python
# Mentioning the heading
fig,ax=plt.subplots(2,2,figsize=(10,10))

# Plot "Age" distribution data
sns.distplot(hr_data['Age'], ax=ax[0,0])
ax[0][0].set_title('Age Distribution')

# Plot "TotalWorkingYears" distribution data
sns.distplot(hr_data['TotalWorkingYears'],ax=ax[0][1])
ax[0][1].set_title('Total Working Years')

# Plot "YearsAtCompany" distribution data
sns.distplot(hr_data['YearsAtCompany'],ax=ax[1][0])
ax[1][0].set_title('Years At Company')

# Plot "YearsInCurrentRole" distribution data
sns.distplot(hr_data['YearsInCurrentRole'],ax=ax[1][1])
ax[1][1].set_title('Years In Current Role')

plt.show()
```

```python
# Count Plot of "Attrition"
plt.figure(figsize=(10,3))
sns.countplot(hr_data['Attrition'])
plt.show()
```

```python
# Bar Plot of "Attrition" according to "Monthly Income"
plt.figure(figsize=(4,4))
sns.barplot(x='Attrition',y='MonthlyIncome',data=hr_data,hue='Attrition')
plt.show()
```

```python
# Bar Plot of "Attrition" according to "Monthly Income" based on "Gender"
sns.barplot(x='Attrition',y='MonthlyIncome',hue='Gender',data=hr_data,estimator=np.median)
plt.show()
```

```python
# Bar Plot of "Attrition" according to "Monthly Income" based on "Jobsatisfaction"
sns.barplot(x='Attrition',y='MonthlyIncome',hue='JobSatisfaction',data=hr_data,estimator =np.median)
plt.show()
```

```python
# Bar Plot of "Attrition" according to "Monthly Income" based on "Maritial Status"
sns.barplot(x='Attrition',y='MonthlyIncome',hue='MaritalStatus',data=hr_data,estimator=np.median)
plt.show()
```

```python
# Bar Plot of "Job Satisfaction" according to "Monthly Income"
plt.figure(figsize=(4,4))
sns.barplot(x='JobSatisfaction',y='MonthlyIncome',data=hr_data,hue='JobSatisfaction')
plt.show()
```

```python
# Box Plot of "Job Satisfaction" according to "Monthly Income"
plt.figure(figsize=(4,4))
sns.boxplot(x='JobSatisfaction',y='MonthlyIncome',data=hr_data,hue='JobSatisfaction')
plt.show()
```

```python
# Box Plot of "Job Satisfaction" according to "Monthly Income" based on "Job Level"
plt.figure(figsize=(8,5))
sns.boxplot(x='JobSatisfaction',y='MonthlyIncome',data=hr_data,hue='JobLevel',palette='Set3')
plt.legend(loc='best')
plt.show()
```

```python
# Violin Plot of "Job Satisfaction" according to "Monthly Income"
plt.figure(figsize=(4,4))
sns.violinplot(x='JobSatisfaction',y='MonthlyIncome',data=hr_data,hue='JobSatisfaction')
plt.show()
```

```python
# Violin Plot of "Attrition" according to "Years At Company" based on "Gender"
plt.figure(figsize=(8,5))
sns.violinplot(x='Attrition',y='YearsAtCompany',data=hr_data,hue='Gender',palette="muted",split=True,inner="stick")
plt.show()
```

```python
# Joint Plot of "Age" according to "Monthly Income"
sns.jointplot(x='Age',y='MonthlyIncome',data=hr_data,kind='scatter',color='m')
plt.show()
```

```python
# Catplot
hr_data['age_group'] = pd.cut(hr_data['Age'],3,labels=['Young','Middle','Senior'])
sns.catplot(x='Attrition', y='MonthlyIncome', hue='JobLevel', col = 'age_group', col_wrap=2, data=hr_data, kind='box')
plt.show()
```

```python
# FacetGrid
g= sns.FacetGrid(hr_data, col="JobSatisfaction",row="Gender")
g.map(sns.kdeplot,"MonthlyIncome", "YearsInCurrentRole")
plt.show()
```

```python
# Pairplot
data1=['Attrition','Age','MonthlyIncome','DistanceFromHome']
sns.pairplot(hr_data[data1],kind='reg',diag_kind="kde",hue='Attrition')
plt.show()
```

```python
data2 = ['Gender','HourlyRate','MonthlyRate','PercentSalaryHike']
sns.pairplot(hr_data[data2],kind='reg',diag_kind="kde",hue='Gender')
plt.show()
```

```python
# Select only numeric columns
numeric_columns = hr_data.select_dtypes(include=['float64', 'int64'])

# Calculate correlations
corr = numeric_columns.corr()

# Plot the HeatMap
plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=True, linewidth=.4, fmt='.1f')
plt.show()
```
